## Expresiones regulares

```python
text = "id: 123\nHola mundo\nid: 564\nQue bien\nid: 321\nMe gusta"

open("datos.txt", "w").write(text)
```

```python
import re

pattern = "id:\s(\d+)"

text = open("datos.txt").read()

for m in re.finditer(pattern, text):
    i = m.start(1)
    a = m.group(1)
    j = i + len(a)
    print("{}:{} -> {}".format(i, j, a))
```
